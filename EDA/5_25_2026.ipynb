Selamat datang di *Pembinaan Prodigi*! Pada sesi ini, kita akan belajar cara mengenali dan menganalisis data kita langkah demi langkah.

**1. Persiapan dan Membaca Data**
Langkah pertama adalah mengimpor *library* (alat bantu) dan memasukkan data CSV kita ke dalam variabel bernama `df` (kependekan dari *DataFrame*).

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Mengatur tampilan grafik agar lebih rapi
sns.set_theme(style="whitegrid")

# Load dataset (Ganti 'nama_dataset.csv' dengan nama file Anda yang sebenarnya)
df = pd.read_csv('dataset_sampel_loan_payback.csv')

# Menampilkan 5 baris teratas untuk memastikan data berhasil masuk
df.head()
```

**2. Cek Informasi Dasar Dataset**
Sebelum menggambar grafik, kita harus tahu ukuran data kita (baris x kolom), tipe data tiap kolom (angka atau teks), dan statistik dasarnya (rata-rata, nilai minimum, maksimum).

```python
# Melihat jumlah baris dan kolom
print("Ukuran data (baris, kolom):", df.shape)

print("\n--- Informasi Tipe Data ---")
df.info()

print("\n--- Statistik Dasar Kolom Angka ---")
display(df.describe())
```

**3. Cek Data Kosong (Missing Values) dan Duplikat**
Data di dunia nyata sering kali tidak sempurna. Mari kita periksa apakah ada sel yang kosong atau baris yang datanya dobel.

```python
print("Jumlah data kosong di tiap kolom:")
display(df.isnull().sum())

print("\nJumlah baris yang duplikat persis:")
print(df.duplicated().sum())
```

**4. Analisis Univariat: Histogram (Melihat Sebaran Data)**
Histogram membantu kita melihat bagaimana data angka tersebar. Misalnya, apakah kebanyakan nilai mengumpul di tengah, atau condong ke kiri/kanan?
*(Catatan: Ganti tulisan `Nama_Kolom_Angka` dengan nama kolom sebenarnya dari data Anda, contoh: `Umur` atau `Harga`)*

```python
plt.figure(figsize=(8, 5))

# Ganti 'Nama_Kolom_Angka' dengan nama kolom dari dataset Anda
sns.histplot(data=df, x='loan_paid_back', kde=False, color='blue')

plt.title('Distribusi Data')
plt.show()
```

**5. Analisis Univariat: Boxplot (Mendeteksi Outlier/Data Pencilan)**
Boxplot sangat berguna untuk melihat rentang data normal dan mendeteksi titik-titik di luar batas wajar (outlier) yang digambarkan sebagai titik-titik kecil di luar kotak.

```python
plt.figure(figsize=(8, 4))

# Ganti 'Nama_Kolom_Angka' dengan nama kolom dari dataset Anda
sns.boxplot(data=df, x='loan_amount', color='lightgreen')

plt.title('Boxplot untuk Deteksi Outlier')
plt.show()
```

**6. Analisis Univariat: Bar Chart (Untuk Data Kategori/Teks)**
Bagaimana jika kolomnya berisi teks atau kategori (misalnya: 'Gender' atau 'Status')? Kita bisa menghitung jumlah masing-masing kategori menggunakan grafik batang (*bar chart*).

```python
plt.figure(figsize=(8, 5))

# Ganti 'Nama_Kolom_Kategori' dengan kolom tipe teks (misal: 'Kota' atau 'Gender')
sns.countplot(data=df, x='gender', palette='viridis')

plt.title('Jumlah Data per Kategori')
plt.xticks(rotation=45) # Memutar label bawah agar tidak bertumpuk
plt.show()
```

**7. Analisis Bivariat: Scatter Plot (Hubungan 2 Kolom Angka)**
Scatter plot digunakan untuk melihat apakah ada hubungan (korelasi) antara dua kolom angka. Misalnya, apakah jika nilai A naik, nilai B juga ikut naik?

```python
plt.figure(figsize=(8, 5))

# Ganti x dan y dengan dua kolom angka yang ingin dibandingkan
sns.scatterplot(data=df, x='loan_amount', y='annual_income', alpha=0.6)

plt.title('Hubungan antara Dua Variabel Angka')
plt.show()
```

**8. Analisis Bivariat: Kategori vs Angka**
Bagaimana perbandingan angka jika dibagi berdasarkan kelompok/kategori tertentu? Misalnya, kita ingin melihat perbandingan `Gaji` (Angka) berdasarkan `Divisi Pekerjaan` (Kategori).

```python
plt.figure(figsize=(10, 6))

# x untuk kolom kategori, y untuk kolom angka
sns.boxplot(data=df, x='gender', y='loan_amount', palette='viridis')

plt.title('Perbandingan Angka Berdasarkan Kategori')
plt.xticks(rotation=45)
plt.show()
```

**9. Analisis Keseluruhan: Heatmap Korelasi**
Terakhir, kita bisa melihat rangkuman hubungan (korelasi) antar **semua** kolom angka sekaligus. Nilai mendekati 1 atau -1 berarti hubungannya sangat kuat. Nilai mendekati 0 berarti tidak ada hubungan.

```python
plt.figure(figsize=(10, 8))

# Memilih hanya kolom angka untuk dihitung korelasinya
kolom_angka = df.select_dtypes(include=['int64', 'float64'])

# Membuat heatmap
sns.heatmap(kolom_angka.corr(), annot=True, cmap='coolwarm', fmt=".2f")

plt.title('Peta Korelasi antar Kolom Angka')
plt.show()
```
